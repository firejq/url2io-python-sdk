```python
# URL2io Python SDK

This is the URL2io python SDK suite. Note that python2.7 is required.

[API Docs](http://www.url2io.com/docs), [URL2io Python SDK下载地址](https://github.com/url2io/url2io-python-sdk)

## url2io.py

This is the underlying API implementation.

###How to use? 
```

```python
import url2io

url2io.API.server += 'demo/' # 此为演示需要，实际使用时请勿添加
```

###Step1. [Register](http://www.url2io.com/accounts/register) and get **`token`**

```python
api = url2io.API('demotoken') # 替换为你注册后得到的token
```

###Step2. Usage

###使用方式 1：`content`字段为网页的正文的html格式，无`next`字段(下一页地址)

```python
ret = api.article(url='http://tech.sina.com.cn/i/2010-08-18/19554560539.shtml')

ret # SDK内部会对返回结果进行解码，使得返回的字段是unicode的
```

###使用方式 2：`content`字段为网页的正文的html格式，含`next`字段(下一页地址)

```python
ret = api.article(url='http://tech.sina.com.cn/i/2010-08-18/19554560539.shtml', fields=['next',])

ret # SDK内部会对返回结果进行解码，使得返回的字段是unicode的
```

###使用方式 3：`content`字段为网页的正文的纯文本，含`next`字段(下一页地址)

```python
ret = api.article(url='http://tech.sina.com.cn/i/2010-08-18/19554560539.shtml', fields=['text', 'next'])

ret # SDK内部会对返回结果进行解码，使得返回的字段是unicode的
```

```python
import json

# 转换成json
print json.dumps(ret, indent=4, ensure_ascii=False).encode('utf-8')
```
